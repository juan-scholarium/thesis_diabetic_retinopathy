```python
import cv2
import numpy as np
import os
import fnmatch
from matplotlib import pyplot as plt
from skimage.exposure import rescale_intensity
```

```python
dir_src = "D:/Datasets/thesis_dr/aptos2019/train_images_300/train_images_rescaled_300_enhanced_green_crop_nakhon/"
dir_dst = "D:/Datasets/thesis_dr/aptos2019/train_images_300/rescaled_enhanced_green_rgb/"
```

```python
fnames = fnmatch.filter(os.listdir(dir_src), "*.png")
fnames[:3]
```

```python
fpath = os.path.join(dir_src, fnames[0])
im = cv2.imread(fpath)
plt.matshow(im)
```

```python
im.shape
```

```python
im_g = im[:, :, 1]
plt.imshow(im_g)
```

```python
im_r = im_b = np.zeros(im_g.shape)
```

```python
im_rgb = rescale_intensity(np.dstack((im_r, im_g, im_b)), in_range=(0, 255))
im_rgb = (im_rgb * 255).astype("uint8")
```

```python
plt.imshow(im_rgb)
```

```python
counter = 0
list_error = list()

for f in fnames:

    try:
        counter += 1
        if counter % 250 ==0:
            print("Image processed : ", counter)

        fpath = os.path.join(dir_src, f)
        im = cv2.imread(fpath)
        im_g = im[:, :, 1]
        im_r = im_b = np.zeros(im_g.shape)
        im_rgb = np.dstack((im_r, im_g, im_b))
        cv2.imwrite(os.path.join(dir_dst, f), im_rgb)
    except:
        list_error.append(f)
    
```
