```python
from skimage import io
import os
from matplotlib import pyplot as plt
```

```python
dir_jpeg = 'D:/Datasets/thesis_dr/aptos2019/train_images_300/train_images_rescaled_300_jpeg_72/'
dir_png = 'D:/Datasets/thesis_dr/aptos2019/train_images_300/train_images_rescaled_300/'
```

```python
fnames_jpeg = os.listdir(dir_jpeg)
fnames_png = os.listdir(dir_png)
```

```python
im_jpeg = io.imread(os.path.join(dir_jpeg, fnames_jpeg[0]))
im_png = io.imread(os.path.join(dir_png, fnames_png[0]))
```

```python
io.imshow(im_jpeg)
```

```python
io.imshow(im_png)
```

```python
im_delta = im_png - im_jpeg
```

```python
io.imshow(im_delta)
```

```python
list_lbl = [
    'Citra PNG', 'Citra JPEG kualitas 72', 'Selisih citra PNG dan JPEG'
]
list_im = [im_png, im_jpeg, im_delta]

plt.figure(figsize=(15, 5))

for idx in range(3):
    plt.subplot(1, 3, idx + 1)
    plt.imshow(list_im[idx])
    plt.title(str(list_lbl[idx]))
```
